```python
import numpy as np

class LinearRegression:

    def __init__(self, x, y):
        self.vetorX = x
        self.vetorY = y
        self.__coeficiente_de_correlacao = self.__correlacao()
        self.__inclinacao = self.__inclinacao()
        self.__interceptacao = self.__interceptacao()

    def __correlacao(self):
        covariacao = np.cov(self.vetorX, self.vetorY, bias=True)[0][1]
        varianciaX = np.var(self.vetorX)
        varianciaY = np.var(self.vetorY)
        return covariacao / np.sqrt(varianciaX * varianciaY)

    def __inclinacao(self):
        desvioPadraoX = np.std(self.vetorX)
        desvioPadraoY = np.std(self.vetorY)
        return self.__coeficiente_de_correlacao * (desvioPadraoY / desvioPadraoX)

    def __interceptacao(self):
        mediaX = np.mean(self.vetorX)
        mediaY = np.mean(self.vetorY)
        return mediaY - self.__inclinacao * mediaX

    def previsao(self, valor):
        return self.__interceptacao + (self.__inclinacao * valor)
    
```

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

regressaoLinear = LinearRegression(x, y)
print(regressaoLinear.previsao(12))
```
