```python
import pandas as pd # Carregar e limpar os dados
import numpy as np  # fazer cálculos matemáticos
import matplotlib.pyplot as plt # Visualizar os dados
import statsmodels.formula.api as sm # Modelagem estatística
import scipy.stats as stats # Testes estatísticos
import seaborn as sns # Visualização de dados
```

```python
dadosBase = pd.read_csv('mt_cars.csv') # Carregar os dados do arquivo CSV
dadosBase.shape # Verificar o número de linhas e colunas do DataFrame
```

```python
dadosBase.head() # Visualizar as primeiras linhas do DataFrame
```

```python
dadosBase = dadosBase.drop(['Unnamed: 0'], axis=1) # Remover a coluna dos nomes dos carros
```

```python
correlacao = dadosBase.corr() # Calcular a correlação entre as variáveis
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f') # Visualizar a matriz de correlação
```

```python
coluna_pares = [('mpg', 'cyl'), ('mpg', 'disp'), ('mpg', 'hp'), ('mpg', 'wt'), ('mpg', 'drat'), ('mpg', 'vs')] # Definir os pares de colunas para os gráficos de dispersão
numero_graficos = len(coluna_pares) # Número de gráficos a serem criados
fig, axes = plt.subplots(nrows=numero_graficos, ncols=1, figsize=(6, 4 * numero_graficos)) # cria uma grade de gráficos

for i, pair in enumerate(coluna_pares): # desenha cada gráfico de dispersão para cada par de colunas
    x_col, y_col = pair # desempacota o par de colunas
    sns.scatterplot(x=x_col, y=y_col, data=dadosBase, ax=axes[i]) # desenha o gráfico de dispersão usando seaborn
    axes[i].set_title(f'{x_col} vs {y_col}') # adiciona um título ao gráfico

plt.tight_layout()
plt.show()
```
